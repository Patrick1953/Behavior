```python
import time, sys

from Interface_executions import Interface_executions

from Lanceur_alimentation import Lanceur_alimentation
from Lanceur_apprentissage import Lanceur_apprentissage
from Lanceur_execution import Lanceur_execution

path = "../outils"
if path not in sys.path :
    sys.path.append (path)
from Entree_sortie_lock import Entree_sortie_lock


from pprint import PrettyPrinter 
def P (stuff , pp = PrettyPrinter(indent=4)) :
    return pp.pprint(stuff)

class bcolors:
    OK = '\033[92m' #GREEN
    WARNING = '\033[93m' #YELLOW
    FAIL = '\033[91m' #RED
    RESET = '\033[0m' #RESET COLOR
    
def controle_running (lanceur, arg) :
    debut = time.time ()
    
    message = f"{bcolors.OK}Are you sure you want to continue? (y/N) {bcolors.RESET}"
    print ()
    P (arg)
    while (True) :
        rep1 = input(message)
        if rep1 == 'N' :
            break
        if rep1 != 'y':
            continue
        t0 = time.time ()
        job = lanceur (arg)
        resultat = True
        while (resultat) :
            print (f"{bcolors.OK}.{bcolors.RESET}", end = "")
            time.sleep (0.1)
            resultat = job.is_alive ()
            continue
        break
    
    print ()
    try :
        x = job
    except: 
        return
    
    if job.get_exitcode () == 0 :
        print (f"{bcolors.OK}Successful in " + str(time.time() - t0)+ f" secondes {bcolors.RESET}")
    else:
        print (f"{bcolors.FAIL}##### Error {bcolors.RESET}")
        
def affichage_resultat (nom_data1, nom_data2) :
    pathFile_evenements = 'dico_evenements_2.json'
    path = '../data/' + nom_environnement +'/parametres/'

    arg_entree_sortie_lock = {} 

    arg_entree_sortie_lock['pathFile'] = path + pathFile_evenements
    Entree_sortie_evenements = Entree_sortie_lock (arg_entree_sortie_lock)
    dico_evenements = Entree_sortie_evenements.lire_with_lock()
    Entree_sortie_evenements.unlock_lire()
    
    #P (dico_evenements)
    
    P(dico_evenements[nom_data1] [nom_data2])
    
    
    
    
        
            
        
        
     
    
nom_environnement = '#test'   
```

# Alimentation

```python
from Interface_executions import Interface_executions

arg = {}
arg ['nom_environnement'] = nom_environnement
arg ['isPurge_existing_index'] = True
arg ['nom_fichier'] =  'evenements.txt'
arg ['nom_tache_alimentation'] = 'test'

I = Interface_executions ()
liste = I.get_liste_fichiers (nom_environnement)

for i in range (0, len(liste)) :
    print (str(i)+ ' : ' + liste [i])
    continue
    
message = f"{bcolors.OK}Choise of file ? {bcolors.RESET}"

while (True) :
    rep1 = input(message)
    try:
        indice = int(rep1)
    except:
        continue
    if indice < 0 or indice >= len(liste):
        continue
    break

arg ['nom_fichier'] =  liste [indice]
    
message = f"{bcolors.OK}³Erase all data before alimentation (y/N) ? {bcolors.RESET}"

while (True) :
    rep1 = input(message)
    if rep1 == 'y' :
        arg ['isPurge_existing_index'] = True
        break
    if rep1 == 'N' :
        arg ['isPurge_existing_index'] = True
        break
        
    continue

lanceur = controle_running (Lanceur_alimentation, arg)
print ()
affichage_resultat ("alimentation", 'execution')
```
