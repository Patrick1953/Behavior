```python
# coding: utf-8
import json, sys
import streamlit as st

path = "../outils"
if path not in sys.path :
    sys.path.append (path)
from Interface_parametres import Interface_parametres
from page1 execution import page1

def page0 (mon_instance, dico) :
    
    st.set_page_config(
                        page_title="Variables parameters",
                        page_icon="🧊",
                        layout="wide",
                        initial_sidebar_state="expanded",
                        )

    hide_streamlit_style = """
    <style>
    #MainMenu {visibility: hidden;}
    footer {visibility: hidden;}
    </style>

    """

    st.markdown(hide_streamlit_style, unsafe_allow_html=True)
    
    
    def display_message (message) :
        st.write (message)
        return
    
    @st.cache (allow_output_mutation = True)
    def get_image (path = "./image/4x/Fichier 1@4x.png" ) :
        f = open (path, "rb" )
        image = f.read()
        f.close()
        return image
    
    @st.cache (allow_output_mutation = True)
    def get_liste_environnement ():
        
        liste = mon_instance.get_liste_environnements ()
        resultat = ['']
        resultat.extend (liste)
        return liste
    @st.cache (allow_output_mutation = True)
    def get_dico (nom_environnement):
        dico = mon_instance.get_dico_evenements (nom_environnement)
        return dico
    
    
    
    
    
    
    
    image = get_image()
    
    st.title ('Parameters management')
    
    st.write ()
    agree0 = st.checkbox('Next Page', key ='Next_page0' )
    if agree0:
        page1(mon_instance, dico)
        st.stop()
    
    
    
    
    
    
    st.sidebar.image(image)
    
    liste = get_liste_environnement ()
    value = st.sidebar.selectbox("Environnement choice :", liste,)
    
    if value != '' :
        dico = get_dico (value)
        display_message (value)
        value = ''
    
    
    
    display_message ('')
    col1, col2 = st.beta_columns(2)
    with col1:
        st.header("Variables")
        st.json(dico ['position'])

    with col2:
        st.header("Variables type")
        st.json(dico ['type'])
        
    return
    
    
    
    
    
    
    
    
if __name__ == "__main__":
    # execute only if run as a script
    mon_instance = Interface_parametres ()
    dico = mon_instace.get_dico_new ()
    page0 (mon_instance, dico)
```
