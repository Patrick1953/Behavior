```python
from Kernel_BE import Kernel

class Alimentation_bloc () :
    """
    def __init__ (self,
                  index_data,
                  nom_variable,
                  ID_reference_min, # string
                  ID_reference_max, # string
                  taille_bloc, # int
                  arg_kernel,
                  isValeur = True,
                  isID = False,
                  isTrace = False,
                  )
        """
    def __init__ (self, arg) :
        self.index_data = arg ['index_data']
        self.nom_variable = arg ['nom_variable']
        self.ID_reference_min = int (arg['ID_reference_min'])
        self.ID_reference_max = int (arg['ID_reference_max'])
        self.taille_bloc = arg ['taille_bloc']
        self.isValeur = arg['isValeur']
        self.isID = arg['isID']
        self.isTrace = arg ['isTrace']
        
        arg_kernel = arg ['arg_kernel']
        self.ID_reference_base = arg_kernel ['_ID_reference_base']
        
        self.kernel = Kernel(arg_kernel)
        
    def alimentation_bloc (self,p__ID,):
        """
        fournit un dico des dico ( chacun  represente un evenement ou ligne )
        par ID ou non (fonction du parametre self.isID )
        
        lit de ID_reference du debut, à fin en bloc (taille = self.bloc_taille (paramete systeme)
        
        """
        _liste_total = []
        _taille = len(self.ID_reference_base)        
        taille = len(self.ID_reference_base) # taille du string ( valeur max des evenements)
        #print ("type (self.ID_reference_base)", type(self.ID_reference_base))
        #print ("self.ID_reference_base", self.ID_reference_base)
        #print ("taille =", taille)
        _isFirst = self.isTrace
        for _debut_bloc in range(self.ID_reference_min,
                                self.ID_reference_max ,
                                self.taille_bloc ) :
            
            _fin_bloc = _debut_bloc + self.taille_bloc
            if _fin_bloc > self.ID_reference_max :
                _fin_bloc = self.ID_reference_max 

            #print ("self.ID_reference_base + str(debut_bloc) = ", self.ID_reference_base + str(debut_bloc) )
            _debut_bloc_string = (self.ID_reference_base + str(_debut_bloc)) [_taille:]
            _fin_bloc_string = (self.ID_reference_base + str(_fin_bloc)) [_taille:]
            _isFirst = self.isTrace
            if _isFirst :
                print ("1 ________________________________________________________________")
                print ("debut_bloc = ", _debut_bloc)
                print ("fin_bloc_string =", _fin_bloc )
                print ("debut_bloc_string = ", _debut_bloc_string)
                print ("fin_bloc_string =", _fin_bloc_string )
                print ("self.nom_variable =",self.nom_variable)
                print ("self.isValeur =", self.isValeur)
                print ("self.taille_bloc =", self.taille_bloc)

            taille_lu, hits = self.kernel.search_par_bloc (self.index_data,
                                                           p__ID,
                                                           #self.nom_variable,
                                                          _debut_bloc_string,
                                                          _fin_bloc_string,
                                                          size = self.taille_bloc,
                                                          #isValeur = self.isValeur,
                                                          isID = self.isID,
                                                         )
            
            
            #reconstitution de l'evenement par ID_reference => ligne ... evenement
            # et mise au fur et à mesure dans dico_globale par ID_reference
            if len(hits) == 0  :
                print ('ID recherche = ', p__ID)
                continue
                
            _liste_global = []
            
            for _hit in hits :
                _variable_lu = _hit ['_source']
                _nom_variable = _variable_lu ['nom_variable']
                _ID_reference = _variable_lu  ['ID_reference']
                try :
                    _evenement = _dico_global [_ID_reference]
                except:
                    _ID = _variable_lu ['ID']
                    _evenement = {'ID' : _ID}
                    
                _evenement [_nom_variable] = _variable_lu ['valeur']
                _liste_global.append (_evenement)
                
                continue
            
            if _isFirst :
                if len(_liste_global) != 0:
                    print ([enreg ['ID'] for enreg in _liste_global ])
                    print ("2 ________________________________________________________________")
                    print ("debut_bloc = ", _debut_bloc)
                    print ("fin_bloc_string =", _fin_bloc )
                    print ("debut_bloc_string = ", _debut_bloc_string)
                    print ("fin_bloc_string =", _fin_bloc_string )
                    print ("self.nom_variable =",self.nom_variable)
                    print ("self.isValeur =", self.isValeur)
                    print ("self.taille_bloc =", self.taille_bloc)
                    _isFirst = False
            
            _liste_total.extend(_liste_global)
            continue
            
        
            
        
        return _liste_total
    
        
        

            
                                
                                
        
        
        
        
```
