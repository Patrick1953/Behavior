```python
# Filename: run_luigi.py
import luigi, os
 
class PrintNumbers(luigi.Task):
 
    def requires(self):
        return []
 
    def output(self):
        return luigi.LocalTarget("numbers_up_to_10.txt")
 
    def run(self):
        with self.output().open('w') as f:
            for i in range(1, 11):
                f.write("{}\n".format(i))
                
class SquaredNumbers(luigi.Task):
 
    def requires(self):
        return [PrintNumbers()]
    
    def output(self):
        return luigi.LocalTarget("squares.txt")
    
 
    def run(self):
        
        with self.input()[0].open() as fin, self.output().open('w') as fout:  
            for line in fin:
                n = int(line.strip())
                out = n * n
                print (out)
                fout.write("{}:{}\n".format(n, out))
        
        
        
        return

                 
if __name__ == '__main__':
    luigi.build([SquaredNumbers()])
    if os.path.exists("numbers_up_to_10.txt") :
        print ("numbers_up_to_10.txt existe")
        os.remove("numbers_up_to_10.txt")
    if os.path.exists("squares.txt") :
        print ("squares.txt existe")
        os.remove("squares.txt")
    
```
